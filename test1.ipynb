```python
def timeSelect(Usertime=""):
    """ 時間選擇轉換 """
    # 預設回傳 8:00
    if Usertime == "":
        return "8:00"
    try:
        hour = int(Usertime.split(':')[0])
        minute = int(Usertime.split(':')[1])
        print("輸入時間:", hour, minute)
    except ValueError as e:
        print("時間格式錯誤:", e)
        return "8:00"
    if hour < 0 or hour > 24 or minute < 0 or minute > 60:
        print("時間選擇超出範圍")
        return "8:00"
    for h in range(0, 25):
        if h == hour:
            if minute < 30:
                return f"{h:02d}:00"
            elif minute > 30:
                return f"{h:02d}:30"
    return "8:00"

print(timeSelect("8:20"))
print(timeSelect("14:40"))
```

```python
import pytesseract, cv2, numpy as np, re
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression

def VerificationCode_tesseract_ocr(border, Vcode_path= "", tesseract_path= "") -> None:
    """ 驗證碼辨識 """
    if tesseract_path != "" or Vcode_path != "":
        pytesseract.pytesseract.tesseract_cmd = tesseract_path
        img = cv2.imread(Vcode_path)
    else:
        return print("請提供驗證碼圖片路徑及 tesseract 執行檔路徑")
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    denoise = cv2.fastNlMeansDenoising(gray, h= 45) # 去噪 h 值可調整(越大去噪效果越強，但可能會損失細節)
    shold_img = cv2.threshold(denoise, 127, 255, cv2.THRESH_BINARY_INV)[1]
    imgCopy = shold_img.copy()
    width = imgCopy.shape[1]
    height = imgCopy.shape[0]
    imgCopy[:, border:width - border] = 0
    imgCopy[height * 3 // 5:height, 0:width // 2] = 0
    imgCopy[height * 2 // 5:height, width // 2:width] = 0

    border_data = np.where(imgCopy == 255)
    Y_data = border_data[0]
    samples = Y_data.shape[0]
    X_data = border_data[1].reshape(samples, 1)

    regs = LinearRegression()  #建立線性回歸模型
    feature = PolynomialFeatures(degree=2)  #建立多項式特徵，degree為多項式的階數
    X_ = feature.fit_transform(X_data)  #轉換特徵
    regs.fit(X_, Y_data)  #訓練模型

    # 預測
    newX = np.array([[i] for i in range(width)])
    newX = newX.reshape(newX.shape[0], 1)
    newX_ = feature.fit_transform(newX)
    newY = regs.predict(newX_)

    # 根據預測結果進行圖像校正
    img_regs = np.zeros_like(imgCopy)
    newY = newY.round(0)
    for point in np.column_stack([newY, newX]):
        py = int(point[0])
        px = int(point[1])
        w = 3
        img_regs[py - w:py + w, px] = 255
    # 顯示校正後的圖像
    diff_img = cv2.absdiff(shold_img, img_regs)
    # 再進行一次降噪
    final_img = cv2.fastNlMeansDenoising(diff_img, h= 80)

    # 輸出最終結果
    Vcode = pytesseract.image_to_string(
        final_img, config= ("--psm 7"
                            "-c tessedit"
                            "_char_whitelist=ABCDEFGHIJKLMNOPQRSTUVWXYZabcdefghijklmnopqrstuvwxyz0123456789"
                            " "))
    #規率替換
    key = re.sub(r'[^A-Za-z0-9]', '', Vcode)
    key = key.upper()
    key = re.sub(r'S', '5', key)
    if 'O' in key:
        if key.index('O') == 2:
            key = re.sub(r'O', 'D', key)
        elif key.index('O') == 0:
            key = re.sub(r'O', '6', key)
        elif key.index('O') == 1:
            key = re.sub(r'O', 'D', key)
        else:
            key = re.sub(r'O', '6', key)
    # print("辨識結果:", key)
    if key == "" or len(key) != 4:
        key = "TypeError"
    return key
    # cv2.imshow('Verification Code', final_img)
    # cv2.waitKey(0)
    # cv2.destroyAllWindows()

def VerificationCodeOutput(Vcode_path= "", tesseract_path= "") -> None:
    """ 驗證碼多次辨識輸出最終結果 """
    Vdata = []
    for i in range(2, 14):
        # print(f"border = {i}")
        result = VerificationCode_tesseract_ocr(border= i,
                                Vcode_path= Vcode_path,
                                tesseract_path= tesseract_path)
        # print("辨識結果:", result)
        Vdata.append(result)
    items = []
    for res in Vdata:
        if res not in items:
            items.append(res)
        else:
            continue
    # print("最終辨識結果:", items)
    final_data = []
    for res in Vdata:
        for m in items:
            qu = 0
            if res == m:
                qu = 1
                if final_data == []:
                    final_data.append((m, qu))
                    # print("New item", (m, qu))
                else:
                    for i in range(len(final_data)):
                        if m == final_data[i][0]:
                            int_temp = final_data[i][1]
                            int_temp += 1
                            final_data[i] = (m, int_temp)
                            break
                        else:
                            if i == len(final_data) - 1:
                                final_data.append((m, qu))
                                # print("New item", (m, qu))
                                break
                # print("Updata:", final_data)
    # print("各結果出現次數:", final_data)

    for i in range(len(final_data)):
        for j in range(i + 1, len(final_data)):
            if final_data[j][1] > final_data[i][1]:
                final_data[i], final_data[j] = final_data[j], final_data[i]
    if len(items) > 2:
        if final_data[0][0] != "TypeError":
            # print("最終結果為:", final_data[0][0], "次數:", final_data[0][1])
            return final_data[0][0]
        else:
            # print("最終結果為:", final_data[1][0], "次數:", final_data[1][1])
            return final_data[1][0]
    else:
        # print("最終結果為:", final_data[0][0], "次數:", final_data[0][1])
        return final_data[0][0]
print(VerificationCodeOutput(Vcode_path= 'D:\\Desktop\\Ticket-Grabbing-System\\verificationCode.png',
                                tesseract_path= r'C:\Program Files\Tesseract-OCR\tesseract.exe'))
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait, Select
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import TimeoutException, NoSuchElementException, ElementClickInterceptedException
from bs4 import BeautifulSoup
from datetime import datetime
import json, time, os

import ddddocr
import pytesseract, cv2, numpy as np, re
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression

ticket_url = "https://www.thsrc.com.tw/ArticleContent/5b93cc99-3a1e-4000-a545-aa5dfe375708"
google_url = "https://www.google.com/"
order_url = ""
html_item = ""

TicketUserData = {
    "startStation": "台北",
    "endStation": "嘉義",
    "chooseDate": "2025/11/20",
    "chooseTime": "10:00",
    "OneWayOrRoundTrip": True,  # True: 單程, False: 來回
    "StandardCarOrBusinessCar": True,  # True: 標準車廂, False: 商務車廂
    }

TimeSelectItems = {}
Schedule_data = []
select_item = {}
ticket_information = {}

opt = webdriver.ChromeOptions()
opt.add_experimental_option("prefs", {"profile.default_content_setting_values.notifications": 2})
opt.add_experimental_option("excludeSwitches", ["enable-automation"])
opt.add_experimental_option('useAutomationExtension', False)
opt.add_argument("--disable-blink-features=AutomationControlled")
driver = webdriver.Chrome(options= opt)
# keycode = cv2.waitKey(0)

def GetJsonData(JsonFile= 'None.json') -> dict:
    if not os.path.exists(JsonFile):
        print(f"{JsonFile} 檔案不存在，返回空字典")
        return {}
    with open(JsonFile, 'r') as file:
        data = json.load(file)
    return data

def WriteJsonData(data: dict, JsonFile= 'None.json') -> None:
    if not data:
        print("提供的資料為空，無法寫入 JSON 檔案")
        return
    if not os.path.exists(JsonFile):
        print(f"{JsonFile} 檔案不存在，無法寫入資料")
        return
    with open(JsonFile, 'w') as file:
        json.dump(data, file, indent=4)

def VerificationCode_ddddocr(img_path= "") -> None:
    if not img_path:
        print("請提供圖片路徑")
        return
    ocr = ddddocr.DdddOcr(beta=True)  # 切换为第二套ocr模型
    if not os.path.exists(img_path):
        print("圖片路徑不存在")
        return
    image = open(img_path, "rb").read()
    result = ocr.classification(image).upper()
    return result

def VerificationCode_tesseract_ocr(border, Vcode_path= "", tesseract_path= "") -> None:
    """ 驗證碼辨識 """
    if tesseract_path != "" or Vcode_path != "":
        pytesseract.pytesseract.tesseract_cmd = tesseract_path
        img = cv2.imread(Vcode_path)
    else:
        return print("請提供驗證碼圖片路徑及 tesseract 執行檔路徑")
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    denoise = cv2.fastNlMeansDenoising(gray, h= 45) # 去噪 h 值可調整(越大去噪效果越強，但可能會損失細節)
    shold_img = cv2.threshold(denoise, 127, 255, cv2.THRESH_BINARY_INV)[1]
    imgCopy = shold_img.copy()
    width = imgCopy.shape[1]
    height = imgCopy.shape[0]
    imgCopy[:, border:width - border] = 0
    imgCopy[height * 3 // 5:height, 0:width // 2] = 0
    imgCopy[height * 2 // 5:height, width // 2:width] = 0

    border_data = np.where(imgCopy == 255)
    Y_data = border_data[0]
    samples = Y_data.shape[0]
    X_data = border_data[1].reshape(samples, 1)

    regs = LinearRegression()  #建立線性回歸模型
    feature = PolynomialFeatures(degree=2)  #建立多項式特徵，degree為多項式的階數
    X_ = feature.fit_transform(X_data)  #轉換特徵
    regs.fit(X_, Y_data)  #訓練模型

    # 預測
    newX = np.array([[i] for i in range(width)])
    newX = newX.reshape(newX.shape[0], 1)
    newX_ = feature.fit_transform(newX)
    newY = regs.predict(newX_)

    # 根據預測結果進行圖像校正
    img_regs = np.zeros_like(imgCopy)
    newY = newY.round(0)
    for point in np.column_stack([newY, newX]):
        py = int(point[0])
        px = int(point[1])
        w = 3
        img_regs[py - w:py + w, px] = 255
    # 顯示校正後的圖像
    diff_img = cv2.absdiff(shold_img, img_regs)
    # 再進行一次降噪
    final_img = cv2.fastNlMeansDenoising(diff_img, h= 80)

    # 輸出最終結果
    Vcode = pytesseract.image_to_string(
        final_img, config= ("--psm 7"
                            "-c tessedit"
                            "_char_whitelist=ABCDEFGHIJKLMNOPQRSTUVWXYZabcdefghijklmnopqrstuvwxyz0123456789"
                            " "))
    #規率替換
    key = re.sub(r'[^A-Za-z0-9]', '', Vcode)
    key = key.upper()
    key = re.sub(r'S', '5', key)
    if 'O' in key:
        if key.index('O') == 2:
            key = re.sub(r'O', 'D', key)
    #     elif key.index('O') == 0:
    #         key = re.sub(r'O', '6', key)
    #     elif key.index('O') == 1:
    #         key = re.sub(r'O', 'D', key)
    #     else:
    #         key = re.sub(r'O', '6', key)
    # print("辨識結果:", key)
    if key == "" or len(key) != 4:
        key = "TypeError"
    return key
    # cv2.imshow('Verification Code', final_img)
    # cv2.waitKey(0)
    # cv2.destroyAllWindows()

def VerificationCodeOutput(Mod = "tesseract", Vcode_path= "", tesseract_path= "") -> None:
    """ 驗證碼多次辨識輸出最終結果 """
    Vdata = []
    if Mod == "ddddocr":
        result = VerificationCode_ddddocr(img_path= Vcode_path)
        return result
    elif Mod != "tesseract":
        for i in range(2, 14):
            # print(f"border = {i}")
            result = VerificationCode_tesseract_ocr(border= i,
                                    Vcode_path= Vcode_path,
                                    tesseract_path= tesseract_path)
            # print("辨識結果:", result)
            Vdata.append(result)
    items = []
    for res in Vdata:
        if res not in items:
            items.append(res)
        else:
            continue
    # print("最終辨識結果:", items)
    final_data = []
    for res in Vdata:
        for m in items:
            qu = 0
            if res == m:
                qu = 1
                if final_data == []:
                    final_data.append((m, qu))
                    # print("New item", (m, qu))
                else:
                    for i in range(len(final_data)):
                        if m == final_data[i][0]:
                            int_temp = final_data[i][1]
                            int_temp += 1
                            final_data[i] = (m, int_temp)
                            break
                        else:
                            if i == len(final_data) - 1:
                                final_data.append((m, qu))
                                # print("New item", (m, qu))
                                break
                # print("Updata:", final_data)
    # print("各結果出現次數:", final_data)

    for i in range(len(final_data)):
        for j in range(i + 1, len(final_data)):
            if final_data[j][1] > final_data[i][1]:
                final_data[i], final_data[j] = final_data[j], final_data[i]
    if len(items) > 2:
        if final_data[0][0] != "TypeError":
            # print("最終結果為:", final_data[0][0], "次數:", final_data[0][1])
            return final_data[0][0]
        else:
            # print("最終結果為:", final_data[1][0], "次數:", final_data[1][1])
            return final_data[1][0]
    else:
        # print("最終結果為:", final_data[0][0], "次數:", final_data[0][1])
        return final_data[0][0]

def monthChange(month) -> None:
    month_dict = {
        1: '一月',
        2: '二月',
        3: '三月',
        4: '四月',
        5: '五月',
        6: '六月',
        7: '七月',
        8: '八月',
        9: '九月',
        10: '十月',
        11: '十一月',
        12: '十二月'
    }
    return month_dict.get(month, "")

def timeSelect(Usertime="") -> None:
    """ 時間選擇轉換 """
    # 預設回傳 08:00
    if Usertime == "":
        return "08:00"
    try:
        t_item = datetime.strptime(Usertime, "%H:%M")
        print("輸入時間:", Usertime)
    except ValueError as e:
        print("時間格式錯誤:", e)
        return "08:00"
    hour = t_item.hour
    minute = t_item.minute
    for h in range(0, 25):
        if h == hour:
            if minute < 30:
                return f"{h:02d}:00"
            elif minute > 30:
                return f"{h:02d}:30"
    return "08:00"

def timechack(TU_da= {}, time_da={}) -> None:
    """ 比較時間字串，回傳 True 或 False """
    # if len(t_day) != 3 or len(u_day) != 3:
    #     return False
    try:
        end_date = datetime.strptime(time_da["timedata"]['limit'], "%Y/%m/%d")
        today_date = datetime.strptime(time_da["timedata"]['date'], "%Y/%m/%d")
        u_date = datetime.strptime(TU_da['chooseDate'], "%Y/%m/%d")
        u_time = datetime.strptime(TU_da['chooseTime'], "%H:%M")
        # print(end_date.month, end_date.day, today_date.month, today_date.day, u_date.month, u_date.day, u_time.hour, u_time.minute)
        if u_date.month >= today_date.month and u_date.month <= end_date.month:
            return True
        if u_date.month == end_date.month:
            if u_date.day <= end_date.day:
                return True
            else:
                return False
        elif u_date.month == today_date.month:
            if u_date.day >= today_date.day:
                return True
            else:
                return False
    except ValueError as e:
        print("時間格式錯誤:", e)

def interOrderInformation(wait, ticketUData={}, timedata={}):
    """ 輸入訂票資訊 """
    """======================"""
    """ 選擇出發站 """
    try:
        start_station = wait.until(EC.element_to_be_clickable(
            (By.XPATH, f"//select[@name='selectStartStation']//option[text()= '{UserData.get('startStation')}']")
        ))
        start_station.click()
        print("選擇出發站:", UserData.get('startStation'))
    except TimeoutException:
        print("沒找到訂票資訊區塊")
    
    """ 選擇終點站 """
    try:
        end_station = wait.until(EC.element_to_be_clickable(
            (By.XPATH, f"//select[@name='selectDestinationStation']//option[text()= '{UserData.get('endStation')}']")
        ))
        end_station.click()
        print("選擇結束站:", UserData.get('endStation'))
    except TimeoutException:
        print("沒找到結束站選項")
    
    """ 選擇日期 """
    try:
        date_input = wait.until(EC.element_to_be_clickable(
            (By.XPATH, "//div[@data-triptype='oneway']")
        ))
        # print(date_input.get_attribute("outerHTML"))
        htmlitem = date_input.get_attribute("outerHTML")
        date_input.click()
        # 解析 HTML 以取得 input 欄位的屬性
        soup = BeautifulSoup(htmlitem, 'html.parser')
        soup_input = soup.find('input', {'name': 'toTimeInputField'})
        timedata.update({"timedata": {}})  # 初始化 timedata 字典，為了寫入Json檔案好閱讀(obj : {...})
        for i in soup_input.attrs:
            timedata["timedata"].update({i: soup_input.attrs[i]})  # 將屬性加入 timedata 字典
        
        if timechack(TU_da= ticketUData, t_da= timedata):
            try:
                U_date = datetime.strptime(ticketUData['chooseDate'], "%Y/%m/%d")
                T_date = datetime.strptime(timedata["timedata"]['date'], "%Y/%m/%d")
                if U_date.month == T_date.month:
                    index = 1
                else:
                    index = 2
                print("選擇日期 index:", index)
                month = monthChange(U_date.month)
                print("選擇月份:", U_date.month, "轉換後:", month)
                day = U_date.day
                print("選擇日期:", day)
                data_chooce = wait.until(EC.element_to_be_clickable(
                    (By.XPATH, f"(//div[contains(@class, 'dayContainer')])[{index}]//span[text()= '{day}' and contains(@aria-label, '{month}') and contains(@class, 'flatpickr-day')]")
                )) # 選擇日期
                data_chooce.click()
                # elems = driver.find_elements(By.XPATH, "//div[contains(@class, 'dayContainer')]")
                # print(f"找到 {len(elems)} 個dayContainer")
                # for e in elems:
                #     print(e.get_attribute("outerHTML"))
            except TimeoutException:
                print("選擇日期失敗，時間超時")
        else:
            print("日期選擇超出範圍或格式錯誤")
    except TimeoutException:
        print("日期選擇區塊沒找到，時間超時")

    """ 選擇時間 """
    try:
        time_item = timeSelect(Usertime= TicketUserData.get('chooseTime', ''))
        time_option = wait.until(EC.element_to_be_clickable(
            (By.XPATH, f"//select[@name='toTimeTable']//option[text()= '{time_item}']")
        ))
        time_option.click()
        print("選擇時間:", time_item)
    except TimeoutException:
        print("沒找到時間選項")

def VerificationCode(driver, wait, Vcode_path= "") -> None:
    """ 處理驗證碼 """
    try:
        """ 驗證碼處理下載 """
        time.sleep(0.5)  # 等待圖片載入
        VCode = wait.until(EC.presence_of_element_located(
        (By.CSS_SELECTOR, "img#BookingS1Form_homeCaptcha_passCode.captcha-img")
        ))
        VCode.screenshot(Vcode_path) # 儲存圖片
        """ 驗證碼辨識 """
        time.sleep(0.5)  # 等待以確保圖片已保存
        result = VerificationCodeOutput(
            Mod= "ddddocr",
            Vcode_path= Vcode_path,
            tesseract_path= r'C:\Program Files\Tesseract-OCR\tesseract.exe'
        )
        if result == "TypeError":
            print("驗證碼辨識失敗，重新嘗試")
            try_again = wait.until(EC.element_to_be_clickable(
                (By.XPATH, "//span[@title='重新產生' and contains(@class, 'material-icons')]")
            ))
            try_again.click()
            VerificationCode(driver, wait, Vcode_path= Vcode_path)  # 遞迴呼叫自己重新嘗試
            return
        """ 輸入驗證碼 """
        input_Vcode = wait.until(EC.element_to_be_clickable(
            (By.CSS_SELECTOR, "input#securityCode.uk-input")
        ))
        input_Vcode.send_keys(result)
        """ 提交驗證碼 """
        time.sleep(0.5)  # 等待以確保輸入完成
        submit_btn = wait.until(EC.element_to_be_clickable(
            (By.CSS_SELECTOR, "input#SubmitButton.uk-align-center.uk-button.uk-button-primary")
        ))
        submit_btn.click()
        # 檢查網址是否改變以判斷是否成功
        time.sleep(0.5)  # 等待頁面反應
        now_url = driver.current_url
        if "?wicket:interface=:0:" not in now_url:  # 假設成功後網址不包含 "?wicket:interface=:0:"
            print("驗證碼正確，進入下一步")
            return
        else: # 驗證碼錯誤, 重新嘗試
            print("驗證碼錯誤，重新嘗試")
            VerificationCode(driver, wait, Vcode_path= Vcode_path)  # 遞迴呼叫自己重新嘗試
    except TimeoutException:
        print("驗證碼圖片錯誤或載入超時")

def ShiftArrangement(shiftData = []) -> None:
    """ 班次安排 """
    print("優先選擇班次離出發時間最近的班次，車程時間最短的班次")
    for i in range(len(shiftData)):
        for j in range(i + 1, len(shiftData)):
            dep_time_i = datetime.strptime(shiftData[i]['querydeparture'], '%H:%M')
            dep_time_j = datetime.strptime(shiftData[j]['querydeparture'], '%H:%M')
            est_time_i = datetime.strptime(shiftData[i]['queryestimatedtime'], '%H:%M')
            est_time_j = datetime.strptime(shiftData[j]['queryestimatedtime'], '%H:%M')
            # 比較出發時間
            if dep_time_j < dep_time_i:
                shiftData[i], shiftData[j] = shiftData[j], shiftData[i]
            # 比較車程時間
            if est_time_j < est_time_i:
                    shiftData[i], shiftData[j] = shiftData[j], shiftData[i]
    return shiftData[0]

def ShiftSelect(wait, tickerUData= {}, ReserveTime= 0) -> None:
    """ 選擇班次 """
    try:
        for i in range(7):  # 假設最多有 7 個班次可選
            shift_item = wait.until(EC.element_to_be_clickable(
                (By.XPATH, f"(//label[contains(@class, 'uk-flex uk-flex-middle result-item')])[{i+1}]")
            ))
            # print("班次按鈕:", schedule_btn)
            soup = BeautifulSoup(shift_item.get_attribute("outerHTML"), 'html.parser')
            input_information = soup.find('input', {'type': 'radio'})
            start_time = datetime.strptime(input_information.attrs['querydeparture'], "%H:%M")
            choose_time = datetime.strptime(tickerUData['chooseTime'], "%H:%M")
            choose_time = choose_time.replace(minute=choose_time.minute + ReserveTime)
            if start_time <= choose_time:
                continue
            # 比較時間，選擇最接近的班次
            temp_data = {}
            for item in input_information.attrs:
                temp_data.update({item: input_information.attrs[item]})  # 將屬性加入 temp_data 字典
            temp_data.update({"SelectListIndex": i + 1})  # 加入班次索引
            Schedule_data.append(temp_data)
        select_item.update({"select_item": {}})  # 初始化 select_item 字典，為了寫入Json檔案好閱讀(obj : {...})
        select_item["select_item"] = ShiftArrangement(shiftData= Schedule_data)
        # print(f"選擇班次: 發車時間 {select_item['querydeparture']}, 抵達時間 {select_item['queryarrival']}, 車次 {select_item['querycode']}, 車程時間 {select_item['queryestimatedtime']}")
        # 點選班次
        shift_btn = wait.until(EC.element_to_be_clickable(
            (By.XPATH, f"(//label[contains(@class, 'uk-flex uk-flex-middle result-item')])[{select_item['select_item']['SelectListIndex']}]")
        ))
        shift_btn.click()
        # for idx, item in enumerate(Schedule_data):
        #     print(f"班次 {idx + 1}: {item}")
    except TimeoutException:
        print("沒找到班次查詢按鈕")
    
    try:
        """ 取得訂單明細 """
        # 訂單最終檢查
        print("訂單明細")
        ticket_information = {}
        info = wait.until(EC.element_to_be_clickable(
            (By.XPATH, "//div[@class= 'uk-child-width-1-1 uk-child-width-1-2@m uk-grid-small uk-grid uk-grid-stack']")
        ))
        soup = BeautifulSoup(info.get_attribute("outerHTML"), 'html.parser')
        items = soup.find_all('div', {'class': 'uk-child-width-1-1 uk-child-width-1-2@m uk-grid-small uk-grid uk-grid-stack'})
        ticket_card = items[0].find('div', {'class': 'ticket-card'})
        travel_type = ticket_card.find('span', {'class': 'direction'}).span.text
        Date = ticket_card.find('span', {'id': 'InfoDepartureDate0'}).text
        car_id = ticket_card.find('span', {'id': 'InfoCode0'}).text
        start_time = ticket_card.find('span', {'id': 'InfoDeparture0'}).text
        start_stn = ticket_card.find('div', {'class': 'departure uk-flex-none'}).find('p', {'class': 'departure-stn'}).span.text
        travel_time = ticket_card.find('span', {'id': 'InfoEstimatedTime0'}).text
        end_time = ticket_card.find('span', {'id': 'InfoArrival0'}).text
        end_stn = ticket_card.find('div', {'class': 'arrival uk-flex-none'}).find('p', {'class': 'arrival-stn'}).span.text
        other = items[0].find('div', {'class': 'extra-info uk-flex uk-flex-column'})
        car_type = other.find_all('div', {'class': 'uk-flex uk-flex-between'})[0].find('span', {'class': 'info-data'}).text
        ticket_type = other.find_all('div', {'class': 'uk-flex uk-flex-between'})[1].find('span', {'class': 'info-data uk-text-right'}).span.text
        ticket_information.update({"ticket_information": {}}) # 初始化 ticket_information 字典，為了寫入Json檔案好閱讀(obj : {...})
        ticket_information["ticket_information"]["ticket_type"] = ticket_type
        ticket_information["ticket_information"]['Date'] = Date
        ticket_information["ticket_information"]['car_id'] = car_id
        ticket_information["ticket_information"]['start_time'] = start_time
        ticket_information["ticket_information"]['start_stn'] = start_stn
        ticket_information["ticket_information"]['travel_time'] = travel_time
        ticket_information["ticket_information"]['end_time'] = end_time
        ticket_information["ticket_information"]['end_stn'] = end_stn
        ticket_information["ticket_information"]['car_type'] = car_type
        ticket_information["ticket_information"]['travel_type'] = travel_type
        # print(f"車次資訊: {travel_type}, 日期: {Date}, 車次: {car_id}, 出發時間: {start_time}, 出發站: {start_stn}, 車程時間: {travel_time}, 抵達時間: {end_time}, 抵達站: {end_stn}, 車廂類型: {car_type}")
        print("|----------------------訂單明細---------------------|")
        print(f"|{travel_type} {Date}")
        print(f"|{start_stn} -------------------> {end_stn}")
        print(f"|{start_time} ====== {travel_time} ====== {end_time}")
        print(f"|車廂: {car_type} | 票數: {ticket_type}")
        print("|--------------------------------------------------|")
    except TimeoutException:
        print("沒找到票種資訊區塊")
    
    try:
        """ 點選班次確定按鈕 """
        query_btn = wait.until(EC.element_to_be_clickable(
            (By.XPATH, "//input[contains(@name, 'SubmitButton') and contains(@class, 'uk-button uk-button-primary uk-button-large btn-next')]")
        ))
        query_btn.click()
        print("點選班次確定按鈕")
    except TimeoutException:
        print("沒找到班次查詢按鈕")


def TicketInformation(driver, wait, U_Mod = 1, member = 2, U_data = {}) -> None:
    """ 取得票種資訊 """
    wait_2 = WebDriverWait(driver, 2) # 短時間等待 2 秒
    try:
        price = wait.until(EC.element_to_be_clickable(
            (By.XPATH, "//div[@class= 'uk-flex uk-flex-between uk-flex-middle ticket-total']")
        ))
        soup = BeautifulSoup(price.get_attribute("outerHTML"), 'html.parser')
        items = soup.find_all('div', {'class': 'uk-flex uk-flex-between uk-flex-middle ticket-total'})
        price_info = items[0].find('p', {'id': 'TotalPrice'}).text
        print("總價", price_info)
        # ticket_information["ticket_information"]['total_price'] = price_info
    except TimeoutException:
        print("沒找到票種價格")
    print("填寫取票人資訊")
    try:
        if U_Mod == 1:
            path = "//select[@name='idInputRadio']//option[text()= '身分證字號']"
            type_text = "身分證字號"
        elif U_Mod == 2:
            path = "//select[@name='idInputRadio']//option[text()= '護照/居留證/入出境許可證號碼']"
            type_text = "護照/居留證/入出境許可證號碼"
        else:
            path = "//select[@name='idInputRadio']//option[text()= '身分證字號']"
            type_text = "身分證字號"
        Uid_select = wait.until(EC.element_to_be_clickable(
            (By.XPATH, path)
        ))
        Uid_select.click()
        print(f"選擇{type_text}作為身份驗證")
    except TimeoutException:
        print("沒找到身份驗證選項")

    try:
        Uid_input = wait.until(EC.element_to_be_clickable(
            (By.XPATH, "//input[@name='dummyId' and @id='idNumber']")
        ))
        Uid_input.send_keys(U_data.get('UserBaseData').get('UID', ''))
    except TimeoutException:
        print("沒找到身份證輸入欄位")

    try:
        phone_input = wait.until(EC.element_to_be_clickable(
            (By.XPATH, "//input[@name='dummyPhone' and @id='mobilePhone']")
        ))
        phone_input.send_keys(U_data.get('UserBaseData').get('', ''))
    except TimeoutException:
        print("沒找到電話輸入欄位")
    
    try:
        email_input = wait.until(EC.element_to_be_clickable(
            (By.XPATH, "//input[@name='email' and @id='email']")
        ))
        email_input.send_keys(U_data.get('GoogleAccount').get('Email_2', ''))
    except TimeoutException:
        print("沒找到 email 輸入欄位")

    try:
        if U_Mod == 1:
            path = "//select[@class='uk-select id-select superEarlyBird']//option[text()= '身分證字號']"
            type_text = "身分證字號"
        elif U_Mod == 2:
            path = "//select[@class='uk-select id-select superEarlyBird']//option[text()= '護照/居留證/入出境許可證號碼']"
            type_text = "護照/居留證/入出境許可證號碼"
        else:
            path = "//select[@class='uk-select id-select superEarlyBird']//option[text()= '身分證字號']"
            type_text = "身分證字號"
        other_check = wait_2.until(EC.element_to_be_clickable(
            (By.XPATH, path)
        ))
        other_check.click()
        input_id = wait_2.until(EC.element_to_be_clickable(
            (By.XPATH, "//input[@class='uk-input passengerDataIdNumber' and contains(@placeholder, '乘車請攜帶所輸入之證件備查')]")
        ))
        input_id.send_keys(U_data.get('UserBaseData').get('UID', ''))
        print(f"選擇{type_text}作為早鳥身份驗證")
    except TimeoutException:
        print("沒找到早鳥身份驗證選項")
    
    if member == 1:
        try:
            member_check = wait.until(EC.element_to_be_clickable(
                (By.XPATH, "//input[@class='uk-radio' and @id='memberSystemRadio3']")
            ))
            member_check.click()
            print("無會員")
        except TimeoutException:
            print("沒找到無會員選項")
    elif member == 2:
        try:
            member_check = wait.until(EC.element_to_be_clickable(
                (By.XPATH, "//input[@class='uk-radio membership-toggle' and @id='memberSystemRadio1']")
            ))
            member_check.click()
            member_number_input = wait.until(EC.element_to_be_clickable(
                (By.XPATH, "//input[@class='uk-input' and @id='msNumber']")
            ))
            member_number_input.send_keys(U_data.get('UserBaseData').get('callnumber', ''))
            print("有會員")
        except TimeoutException:
            print("沒找到有會員選項")
    elif member == 3:
        try:
            member_check = wait.until(EC.element_to_be_clickable(
                (By.XPATH, "//input[@class='uk-radio membership-toggle' and @id='memberSystemRadio2']")
            ))
            member_check.click()
            member_number_input = wait.until(EC.element_to_be_clickable(
                (By.XPATH, "//input[@class='uk-input' and @id='msNumber']")
            ))
            member_number_input.send_keys(U_data.get('UserBaseData').get('Enterprise_unified_editing', ''))
            print("有企業會員統編")
        except TimeoutException:
            print("沒找到企業會員統編選項")
            
    try:
        Personal_information_btn = wait.until(EC.element_to_be_clickable(
            (By.XPATH, "//input[contains(@name, 'agree') and contains(@class, 'uk-checkbox')]")
        ))
        Personal_information_btn.click()
        print("點選同意個人資料保護聲明")
    except TimeoutException:
        print("沒找到同意個人資料保護聲明選項")
    
    print("填寫取票人資訊完成")
    get_ticket = input("請確認取票人資訊無誤後，是否要送出訂單？(y/n): ")
    if get_ticket.lower() == 'y' or get_ticket.lower() == 'yes':
        try:
            submit_order_btn = wait.until(EC.element_to_be_clickable(
                (By.XPATH, "//input[contains(@id, 'isSubmit') and contains(@value, '完成訂位')]")
            ))
            submit_order_btn.click()
            print("送出訂單")
        except TimeoutException:
            print("沒找到送出訂單按鈕")
    else:
        print("10秒後自動關閉瀏覽器")
        time.sleep(10)
        return

try:
    today = datetime.now().strftime("%Y/%m/%d")
    print("今天日期:", today)
    driver.get(ticket_url)
    driver.set_window_size(width=800, height=1000) # 設定視窗大小(以免元素被遮蔽)
    wait = WebDriverWait(driver, 12) # 可以視網速調整
    try:
        prn_btn = wait.until(EC.element_to_be_clickable(
            (By.CSS_SELECTOR, "button.swal2-confirm")
        ))
        if "我同意" in prn_btn.text:
            prn_btn.click()
        else:
            # 如果 class 找到但不是預期文字，備援走下方逐一比對
            raise NoSuchElementException()
    except (TimeoutException, NoSuchElementException):
        # 二：備援 - 逐一比對所有 button 的文字
        print("嘗試備援：列出所有按鈕文字以供判斷")
        buttons = driver.find_elements(By.TAG_NAME, "button")
        clicked = False
        for b in buttons:
            text = (b.text or "").strip()
            # 列印以利除錯（可註解掉）
            print("button text:", repr(text), "class:", b.get_attribute("class"))
            if text == "我同意" or text == "同意" or "我同意" in text:
                try:
                    wait.until(EC.element_to_be_clickable(
                        (By.XPATH, f"//button[normalize-space(text())='{text}']")
                    ))
                    b.click()
                    clicked = True
                    print("以文字匹配點選按鈕:", text)
                    break
                except ElementClickInterceptedException:
                    # 最後手段：用 JS click
                    driver.execute_script("arguments[0].click();", b)
                    clicked = True
                    print("用 JS click 點選按鈕:", text)
                    break
        if not clicked:
            print("沒找到匹配的「我同意」按鈕（或按鈕不可點）")
    try:
        orderLink = wait.until(EC.element_to_be_clickable(
            (By.CSS_SELECTOR, "li.nav-item a[title*='網路訂票 [另開新視窗]']")
        ))
        orderLink.click()
        order_url = orderLink.get_attribute("href")
        print("訂票連結:", orderLink.get_attribute("href"))
        # 切換到新分頁
        tabs = driver.window_handles
        driver.switch_to.window(tabs[-1]) # 切換到最新開的分頁
        print("標題", driver.title)
    except TimeoutException:
        print("沒找到訂票連結")

    try:
        ok_btn = wait.until(EC.element_to_be_clickable(
            (By.CSS_SELECTOR, "button#cookieAccpetBtn.policy-btn-accept")
        ))
        ok_btn.click()
        print("進入訂票頁面，點選cookie確認按鈕")
    except TimeoutException:
        print("沒找到確認按鈕")
    
    U_da = GetJsonData(JsonFile= 'UserPassword.json')

    interOrderInformation(wait, ticketUData= TicketUserData, timedata= TimeSelectItems)
    VerificationCode(driver, wait, Vcode_path= "D:\\Desktop\\Ticket-Grabbing-System\\verificationCode.png")
    ShiftSelect(wait, tickerUData= TicketUserData, ReserveTime= 10)
    TicketInformation(driver, wait, U_Mod= 1, member= 2, U_data= U_da)
except TimeoutException:
    print("等待元素時發生 Timeout（可能是網頁結構改變或載入較慢）")
finally:
    # if keycode == 27:  # 按下 ESC 鍵
    driver.quit()
    print("browser quit")

# for key, value in TimeSelectItems.items():
#     print(f"{key}: {value}")

if __name__ == "__main__":
    pass

        
```
