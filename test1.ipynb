```python
def timeSelect(Usertime=""):
    """ 時間選擇轉換 """
    # 預設回傳 8:00
    if Usertime == "":
        return "8:00"
    try:
        hour = int(Usertime.split(':')[0])
        minute = int(Usertime.split(':')[1])
        print("輸入時間:", hour, minute)
    except ValueError as e:
        print("時間格式錯誤:", e)
        return "8:00"
    if hour < 0 or hour > 24 or minute < 0 or minute > 60:
        print("時間選擇超出範圍")
        return "8:00"
    for h in range(0, 25):
        if h == hour:
            if minute < 30:
                return f"{h:02d}:00"
            elif minute > 30:
                return f"{h:02d}:30"
    return "8:00"

print(timeSelect("8:20"))
print(timeSelect("14:40"))
```

```python
import pytesseract, cv2, numpy as np, re
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression

def VerificationCodeDecoder(border, Vcode_path= "", tesseract_path= "") -> None:
    """ 驗證碼辨識 """
    if tesseract_path != "" or Vcode_path != "":
        pytesseract.pytesseract.tesseract_cmd = tesseract_path
        img = cv2.imread(Vcode_path)
    else:
        return print("請提供驗證碼圖片路徑及 tesseract 執行檔路徑")
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    denoise = cv2.fastNlMeansDenoising(gray, h= 45) # 去噪 h 值可調整(越大去噪效果越強，但可能會損失細節)
    shold_img = cv2.threshold(denoise, 127, 255, cv2.THRESH_BINARY_INV)[1]
    imgCopy = shold_img.copy()
    width = imgCopy.shape[1]
    height = imgCopy.shape[0]
    imgCopy[:, border:width - border] = 0
    imgCopy[height * 3 // 5:height, 0:width // 2] = 0
    imgCopy[height * 2 // 5:height, width // 2:width] = 0

    border_data = np.where(imgCopy == 255)
    Y_data = border_data[0]
    samples = Y_data.shape[0]
    X_data = border_data[1].reshape(samples, 1)

    regs = LinearRegression()  #建立線性回歸模型
    feature = PolynomialFeatures(degree=2)  #建立多項式特徵，degree為多項式的階數
    X_ = feature.fit_transform(X_data)  #轉換特徵
    regs.fit(X_, Y_data)  #訓練模型

    # 預測
    newX = np.array([[i] for i in range(width)])
    newX = newX.reshape(newX.shape[0], 1)
    newX_ = feature.fit_transform(newX)
    newY = regs.predict(newX_)

    # 根據預測結果進行圖像校正
    img_regs = np.zeros_like(imgCopy)
    newY = newY.round(0)
    for point in np.column_stack([newY, newX]):
        py = int(point[0])
        px = int(point[1])
        w = 3
        img_regs[py - w:py + w, px] = 255
    # 顯示校正後的圖像
    diff_img = cv2.absdiff(shold_img, img_regs)
    # 再進行一次降噪
    final_img = cv2.fastNlMeansDenoising(diff_img, h= 80)

    # 輸出最終結果
    Vcode = pytesseract.image_to_string(
        final_img, config= ("--psm 7"
                            "-c tessedit"
                            "_char_whitelist=ABCDEFGHIJKLMNOPQRSTUVWXYZabcdefghijklmnopqrstuvwxyz0123456789"
                            " "))
    #規率替換
    key = re.sub(r'[^A-Za-z0-9]', '', Vcode)
    key = key.upper()
    key = re.sub(r'S', '5', key)
    if 'O' in key:
        if key.index('O') == 2:
            key = re.sub(r'O', 'D', key)
        elif key.index('O') == 0:
            key = re.sub(r'O', '6', key)
        elif key.index('O') == 1:
            key = re.sub(r'O', 'D', key)
        else:
            key = re.sub(r'O', '6', key)
    # print("辨識結果:", key)
    if key == "" or len(key) != 4:
        key = "TypeError"
    return key
    # cv2.imshow('Verification Code', final_img)
    # cv2.waitKey(0)
    # cv2.destroyAllWindows()

def VerificationCodeOutput_tesseract_ocr(Vcode_path= "", tesseract_path= "") -> None:
    """ 驗證碼多次辨識輸出最終結果 """
    Vdata = []
    for i in range(2, 14):
        # print(f"border = {i}")
        result = VerificationCodeDecoder(border= i,
                                Vcode_path= Vcode_path,
                                tesseract_path= tesseract_path)
        # print("辨識結果:", result)
        Vdata.append(result)
    items = []
    for res in Vdata:
        if res not in items:
            items.append(res)
        else:
            continue
    # print("最終辨識結果:", items)
    final_data = []
    for res in Vdata:
        for m in items:
            qu = 0
            if res == m:
                qu = 1
                if final_data == []:
                    final_data.append((m, qu))
                    # print("New item", (m, qu))
                else:
                    for i in range(len(final_data)):
                        if m == final_data[i][0]:
                            int_temp = final_data[i][1]
                            int_temp += 1
                            final_data[i] = (m, int_temp)
                            break
                        else:
                            if i == len(final_data) - 1:
                                final_data.append((m, qu))
                                # print("New item", (m, qu))
                                break
                # print("Updata:", final_data)
    # print("各結果出現次數:", final_data)

    for i in range(len(final_data)):
        for j in range(i + 1, len(final_data)):
            if final_data[j][1] > final_data[i][1]:
                final_data[i], final_data[j] = final_data[j], final_data[i]
    if len(items) > 2:
        if final_data[0][0] != "TypeError":
            # print("最終結果為:", final_data[0][0], "次數:", final_data[0][1])
            return final_data[0][0]
        else:
            # print("最終結果為:", final_data[1][0], "次數:", final_data[1][1])
            return final_data[1][0]
    else:
        # print("最終結果為:", final_data[0][0], "次數:", final_data[0][1])
        return final_data[0][0]
print(VerificationCodeOutput_tesseract_ocr(Vcode_path= 'D:\\Desktop\\Ticket-Grabbing-System\\verificationCode.png',
                                tesseract_path= r'C:\Program Files\Tesseract-OCR\tesseract.exe'))
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait, Select
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import TimeoutException, NoSuchElementException, ElementClickInterceptedException
from bs4 import BeautifulSoup
import json, time, datetime

import ddddocr
import pytesseract, cv2, numpy as np, re
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression

ticket_url = "https://www.thsrc.com.tw/ArticleContent/5b93cc99-3a1e-4000-a545-aa5dfe375708"
google_url = "https://www.google.com/"
order_url = ""
html_item = ""

UserData = {
    "startStation": "台北",
    "endStation": "嘉義",
    "chooseDate": "2025-11-20",
    "chooseTime": "10:00",
    "OneWayOrRoundTrip": True,  # True: 單程, False: 來回
    "StandardCarOrBusinessCar": True,  # True: 標準車廂, False: 商務車廂
    }

TimeSelectItems = {}

opt = webdriver.ChromeOptions()
opt.add_experimental_option("prefs", {"profile.default_content_setting_values.notifications": 2})
opt.add_experimental_option("excludeSwitches", ["enable-automation"])
opt.add_experimental_option('useAutomationExtension', False)
opt.add_argument("--disable-blink-features=AutomationControlled")
driver = webdriver.Chrome(options= opt)
# keycode = cv2.waitKey(0)

def VerificationCodeOutput_ddddocr(img_path= "") -> None:
    if img_path == "":
        print("請提供圖片路徑")
        return
    ocr = ddddocr.DdddOcr(beta=True)  # 切换为第二套ocr模型
    image = open(img_path, "rb").read()
    result = ocr.classification(image)
    result = result.upper()
    return result

def VerificationCodeDecoder(border, Vcode_path= "", tesseract_path= "") -> None:
    """ 驗證碼辨識 """
    if tesseract_path != "" or Vcode_path != "":
        pytesseract.pytesseract.tesseract_cmd = tesseract_path
        img = cv2.imread(Vcode_path)
    else:
        return print("請提供驗證碼圖片路徑及 tesseract 執行檔路徑")
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    denoise = cv2.fastNlMeansDenoising(gray, h= 45) # 去噪 h 值可調整(越大去噪效果越強，但可能會損失細節)
    shold_img = cv2.threshold(denoise, 127, 255, cv2.THRESH_BINARY_INV)[1]
    imgCopy = shold_img.copy()
    width = imgCopy.shape[1]
    height = imgCopy.shape[0]
    imgCopy[:, border:width - border] = 0
    imgCopy[height * 3 // 5:height, 0:width // 2] = 0
    imgCopy[height * 2 // 5:height, width // 2:width] = 0

    border_data = np.where(imgCopy == 255)
    Y_data = border_data[0]
    samples = Y_data.shape[0]
    X_data = border_data[1].reshape(samples, 1)

    regs = LinearRegression()  #建立線性回歸模型
    feature = PolynomialFeatures(degree=2)  #建立多項式特徵，degree為多項式的階數
    X_ = feature.fit_transform(X_data)  #轉換特徵
    regs.fit(X_, Y_data)  #訓練模型

    # 預測
    newX = np.array([[i] for i in range(width)])
    newX = newX.reshape(newX.shape[0], 1)
    newX_ = feature.fit_transform(newX)
    newY = regs.predict(newX_)

    # 根據預測結果進行圖像校正
    img_regs = np.zeros_like(imgCopy)
    newY = newY.round(0)
    for point in np.column_stack([newY, newX]):
        py = int(point[0])
        px = int(point[1])
        w = 3
        img_regs[py - w:py + w, px] = 255
    # 顯示校正後的圖像
    diff_img = cv2.absdiff(shold_img, img_regs)
    # 再進行一次降噪
    final_img = cv2.fastNlMeansDenoising(diff_img, h= 80)

    # 輸出最終結果
    Vcode = pytesseract.image_to_string(
        final_img, config= ("--psm 7"
                            "-c tessedit"
                            "_char_whitelist=ABCDEFGHIJKLMNOPQRSTUVWXYZabcdefghijklmnopqrstuvwxyz0123456789"
                            " "))
    #規率替換
    key = re.sub(r'[^A-Za-z0-9]', '', Vcode)
    key = key.upper()
    key = re.sub(r'S', '5', key)
    if 'O' in key:
        if key.index('O') == 2:
            key = re.sub(r'O', 'D', key)
    #     elif key.index('O') == 0:
    #         key = re.sub(r'O', '6', key)
    #     elif key.index('O') == 1:
    #         key = re.sub(r'O', 'D', key)
    #     else:
    #         key = re.sub(r'O', '6', key)
    # print("辨識結果:", key)
    if key == "" or len(key) != 4:
        key = "TypeError"
    return key
    # cv2.imshow('Verification Code', final_img)
    # cv2.waitKey(0)
    # cv2.destroyAllWindows()

def VerificationCodeOutput_tesseract_ocr(Vcode_path= "", tesseract_path= "") -> None:
    """ 驗證碼多次辨識輸出最終結果 """
    Vdata = []
    for i in range(2, 14):
        # print(f"border = {i}")
        result = VerificationCodeDecoder(border= i,
                                Vcode_path= Vcode_path,
                                tesseract_path= tesseract_path)
        # print("辨識結果:", result)
        Vdata.append(result)
    items = []
    for res in Vdata:
        if res not in items:
            items.append(res)
        else:
            continue
    # print("最終辨識結果:", items)
    final_data = []
    for res in Vdata:
        for m in items:
            qu = 0
            if res == m:
                qu = 1
                if final_data == []:
                    final_data.append((m, qu))
                    # print("New item", (m, qu))
                else:
                    for i in range(len(final_data)):
                        if m == final_data[i][0]:
                            int_temp = final_data[i][1]
                            int_temp += 1
                            final_data[i] = (m, int_temp)
                            break
                        else:
                            if i == len(final_data) - 1:
                                final_data.append((m, qu))
                                # print("New item", (m, qu))
                                break
                # print("Updata:", final_data)
    # print("各結果出現次數:", final_data)

    for i in range(len(final_data)):
        for j in range(i + 1, len(final_data)):
            if final_data[j][1] > final_data[i][1]:
                final_data[i], final_data[j] = final_data[j], final_data[i]
    if len(items) > 2:
        if final_data[0][0] != "TypeError":
            # print("最終結果為:", final_data[0][0], "次數:", final_data[0][1])
            return final_data[0][0]
        else:
            # print("最終結果為:", final_data[1][0], "次數:", final_data[1][1])
            return final_data[1][0]
    else:
        # print("最終結果為:", final_data[0][0], "次數:", final_data[0][1])
        return final_data[0][0]

def monthChange(month) -> None:
    month_dict = {
        1: '一月',
        2: '二月',
        3: '三月',
        4: '四月',
        5: '五月',
        6: '六月',
        7: '七月',
        8: '八月',
        9: '九月',
        10: '十月',
        11: '十一月',
        12: '十二月'
    }
    return month_dict.get(month, "")

def timeSelect(Usertime="") -> None:
    """ 時間選擇轉換 """
    # 預設回傳 08:00
    if Usertime == "":
        return "08:00"
    try:
        hour = int(Usertime.split(':')[0])
        minute = int(Usertime.split(':')[1])
        print("輸入時間:", hour, minute)
    except ValueError as e:
        print("時間格式錯誤:", e)
        return "08:00"
    if hour < 0 or hour > 24 or minute < 0 or minute > 60:
        print("時間選擇超出範圍")
        return "08:00"
    for h in range(0, 25):
        if h == hour:
            if minute < 30:
                return f"{h:02d}:00"
            elif minute > 30:
                return f"{h:02d}:30"
    return "08:00"

def timechack(us_da= {}, t_da={}) -> None:
    """ 比較時間字串，回傳 True 或 False """
    # if len(t_day) != 3 or len(u_day) != 3:
    #     return False
    try:
        end_month = int(t_da['limit'].split('/')[1])
        end_day = int(t_da['limit'].split('/')[2])
        today_m = int(t_da['date'].split('/')[1])
        today_d = int(t_da['date'].split('/')[2])
        u_month = int(us_da['chooseDate'].split('-')[1])
        u_day = int(us_da['chooseDate'].split('-')[2])
        u_hour = int(us_da['chooseTime'].split(':')[0])
        u_minute = int(us_da['chooseTime'].split(':')[1])
        # print(end_month, end_day, today_m, today_d, u_month, u_day, u_hour, u_minute)
    except ValueError as e:
        print("時間格式錯誤:", e)
        return False
    if u_hour > 0 and u_hour < 24 and u_minute >= 0 and u_minute < 60:
        return True
    else:
        return False
    if u_month >= today_m and u_month <= end_month:
        return True
    else:
        return False
    if u_day >= 1 and u_day <= 31:
        if u_month == end_month:
            if u_day <= end_day:
                return True
            else:
                return False
        elif u_month == today_m:
            if u_day >= today_d:
                return True
            else:
                return False
    else:
        return False

def interOrderInformation(wait, UserData={}, timedata={}):
    """ 輸入訂票資訊 """
    """======================"""
    """ 選擇出發站 """
    try:
        start_station = wait.until(EC.element_to_be_clickable((By.XPATH, f"//select[@name='selectStartStation']//option[text()= '{UserData.get('startStation')}']")))
        start_station.click()
        print("選擇出發站:", UserData.get('startStation'))
    except TimeoutException:
        print("沒找到訂票資訊區塊")
    
    """ 選擇終點站 """
    try:
        end_station = wait.until(EC.element_to_be_clickable((By.XPATH, f"//select[@name='selectDestinationStation']//option[text()= '{UserData.get('endStation')}']")))
        end_station.click()
        print("選擇結束站:", UserData.get('endStation'))
    except TimeoutException:
        print("沒找到結束站選項")
    
    """ 選擇日期 """
    try:
        date_input = wait.until(EC.element_to_be_clickable((By.XPATH, "//div[@data-triptype='oneway']")))
        # print(date_input.get_attribute("outerHTML"))
        htmlitem = date_input.get_attribute("outerHTML")
        date_input.click()
        # 解析 HTML 以取得 input 欄位的屬性
        soup = BeautifulSoup(htmlitem, 'html.parser')
        soup_input = soup.find('input', {'name': 'toTimeInputField'})
        for i in soup_input.attrs:
            timedata.update({i: soup_input.attrs[i]})  # 將屬性加入 timedata 字典
        
        if timechack(us_da= UserData, t_da= timedata):
            try:
                if UserData['chooseDate'].split('-')[1] == timedata['date'].split('/')[1]:
                    index = 1
                else:
                    index = 2
                print("選擇日期 index:", index)
                month = monthChange(int(UserData['chooseDate'].split('-')[1]))
                print("選擇月份:", month)
                day = UserData['chooseDate'].split('-')[2]
                print("選擇日期:", day)
                data_chooce = wait.until(EC.element_to_be_clickable((By.XPATH, f"(//div[contains(@class, 'dayContainer')])[{index}]//span[text()= '{day}' and contains(@aria-label, '{month}') and contains(@class, 'flatpickr-day')]"))) # 選擇日期
                # print(data_chooce.get_attribute("outerHTML"))
                data_chooce.click()
                # elems = driver.find_elements(By.XPATH, "//div[contains(@class, 'dayContainer')]")
                # print(f"找到 {len(elems)} 個dayContainer")
                # for e in elems:
                #     print(e.get_attribute("outerHTML"))
            except TimeoutException:
                print("選擇日期失敗，時間超時")
        else:
            print("日期選擇超出範圍或格式錯誤")
    except TimeoutException:
        print("日期選擇區塊沒找到，時間超時")

    """ 選擇時間 """
    try:
        time_item = timeSelect(Usertime= UserData.get('chooseTime', ''))
        time_option = wait.until(EC.element_to_be_clickable((By.XPATH, f"//select[@name='toTimeTable']//option[text()= '{time_item}']")))
        time_option.click()
        print("選擇時間:", time_item)
    except TimeoutException:
        print("沒找到時間選項")

def VerificationCode(driver, wait):
    """ 處理驗證碼 """
    try:
        """ 驗證碼處理下載 """
        time.sleep(0.5)  # 等待圖片載入
        VCode = wait.until(EC.presence_of_element_located(
        (By.CSS_SELECTOR, "img#BookingS1Form_homeCaptcha_passCode.captcha-img")
        ))
        VCode.screenshot('D:\\Desktop\\Ticket-Grabbing-System\\verificationCode.png') # 儲存圖片
        """ 驗證碼辨識 """
        time.sleep(0.5)  # 等待以確保圖片已保存
        result = VerificationCodeOutput_tesseract_ocr(
            Vcode_path= 'D:\\Desktop\\Ticket-Grabbing-System\\verificationCode.png',
            tesseract_path= r'C:\Program Files\Tesseract-OCR\tesseract.exe'
        )
        # result = VerificationCodeOutput_ddddocr(
        #     img_path= 'D:\\Desktop\\Ticket-Grabbing-System\\verificationCode.png'
        # )
        """ 輸入驗證碼 """
        input_Vcode = wait.until(EC.element_to_be_clickable(
            (By.CSS_SELECTOR, "input#securityCode.uk-input")
        ))
        input_Vcode.send_keys(result)
        """ 提交驗證碼 """
        time.sleep(0.5)  # 等待以確保輸入完成
        submit_btn = wait.until(EC.element_to_be_clickable(
            (By.CSS_SELECTOR, "input#SubmitButton.uk-align-center.uk-button.uk-button-primary")
        ))
        submit_btn.click()
        # 檢查網址是否改變以判斷是否成功
        time.sleep(0.5)  # 等待頁面反應
        now_url = driver.current_url
        if "?wicket:interface=:0:" not in now_url:  # 假設成功後網址不包含 "?wicket:interface=:0:"
            print("驗證碼正確，進入下一步")
            return
        else: # 驗證碼錯誤, 重新嘗試
            print("驗證碼錯誤，重新嘗試")
            # try_again = wait.until(EC.element_to_be_clickable(
            #     (By.XPATH, "//span[@title='重新產生' and contains(@class, 'material-icons')]")
            # ))
            # try_again.click()
            time.sleep(0.5)  # 等待圖片重新載入
            VerificationCode(driver, wait)  # 遞迴呼叫自己重新嘗試
    except TimeoutException:
        print("驗證碼圖片錯誤或載入超時")

def scheduleSelect(driver, wait) -> None:
    """ 選擇班次 """
    try:
        for i in range(5):
            time.sleep(0.5)  # 等待頁面反應
            schedule_btn = wait.until(EC.element_to_be_clickable(
                (By.XPATH, f"(//label[contains(@class, 'uk-flex uk-flex-middle result-item')])[{i+1}]")
            ))
            print("班次按鈕:", schedule_btn)
        # 保存到文件
        # if "班次查詢" in element.text:
        #     element.click()
        print("點選班次查詢按鈕")
    except TimeoutException:
        print("沒找到班次查詢按鈕")

try:
    today = datetime.date.today()
    print("今天日期:", today)
    driver.get(ticket_url)
    driver.set_window_size(width=800, height=1000) # 設定視窗大小(以免元素被遮蔽)
    wait = WebDriverWait(driver, 12) # 可以視網速調整
    try:
        prn_btn = wait.until(EC.element_to_be_clickable((By.CSS_SELECTOR, "button.swal2-confirm")))
        print(prn_btn)
        if "我同意" in prn_btn.text:
            prn_btn.click()
            print("使用 .swal2-confirm 點選")
        else:
            # 如果 class 找到但不是預期文字，備援走下方逐一比對
            raise NoSuchElementException()
    except (TimeoutException, NoSuchElementException):
        # 二：備援 - 逐一比對所有 button 的文字
        print("嘗試備援：列出所有按鈕文字以供判斷")
        buttons = driver.find_elements(By.TAG_NAME, "button")
        clicked = False
        for b in buttons:
            text = (b.text or "").strip()
            # 列印以利除錯（可註解掉）
            print("button text:", repr(text), "class:", b.get_attribute("class"))
            if text == "我同意" or text == "同意" or "我同意" in text:
                try:
                    wait.until(EC.element_to_be_clickable((By.XPATH, f"//button[normalize-space(text())='{text}']")))
                    b.click()
                    clicked = True
                    print("以文字匹配點選按鈕:", text)
                    break
                except ElementClickInterceptedException:
                    # 最後手段：用 JS click
                    driver.execute_script("arguments[0].click();", b)
                    clicked = True
                    print("用 JS click 點選按鈕:", text)
                    break
        if not clicked:
            print("沒找到匹配的「我同意」按鈕（或按鈕不可點）")
    try:
        orderLink = wait.until(EC.element_to_be_clickable((By.CSS_SELECTOR, "li.nav-item a[title*='網路訂票 [另開新視窗]']")))
        orderLink.click()
        order_url = orderLink.get_attribute("href")
        print("訂票連結:", orderLink.get_attribute("href"))
        # 切換到新分頁
        tabs = driver.window_handles
        driver.switch_to.window(tabs[-1]) # 切換到最新開的分頁
        print("標題", driver.title)
    except TimeoutException:
        print("沒找到訂票連結")

    try:
        ok_btn = wait.until(EC.element_to_be_clickable((By.CSS_SELECTOR, "button#cookieAccpetBtn.policy-btn-accept")))
        ok_btn.click()
        print("進入訂票頁面，點選cookie確認按鈕")
    except TimeoutException:
        print("沒找到確認按鈕")
    
    interOrderInformation(wait, UserData= UserData, timedata= TimeSelectItems)
    VerificationCode(driver, wait)
    scheduleSelect(driver, wait)

except TimeoutException:
    print("等待元素時發生 Timeout（可能是網頁結構改變或載入較慢）")
finally:
    # if keycode == 27:  # 按下 ESC 鍵
    # time.sleep(10)
    driver.quit()
    print("browser quit")

# for key, value in TimeSelectItems.items():
#     print(f"{key}: {value}")

if __name__ == "__main__":
    pass

        
```
